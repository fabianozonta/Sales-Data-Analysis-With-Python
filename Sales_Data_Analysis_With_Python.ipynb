Exploratory Data Analisys

```python
# What is the overall sales trend?
df['order_date'].min()
```

```python
df['order_date'].max()
```

```python
# Getting month year from dataset
df['month_year'] = df['order_date'].apply(lambda x: x.strftime('%Y-%m'))
```

```python
df['month_year'] 
```

```python
# Grouping month year
df_trend = df.groupby('month_year').sum()['sales'].reset_index()
```

```python
# Setting the figure size
plt.figure(figsize=(15, 6))
plt.plot(df_trend['month_year'], df_trend['sales'])
plt.xticks(rotation='vertical', size=8, color='red')
plt.show()
```

```python
# What is the top 10 products by sales?

# Grouping product name column
pd.DataFrame(df.groupby('product_name').sum()['sales'])
```

```python
prod_sales = pd.DataFrame(df.groupby('product_name').sum()['sales'])
```

```python
# Sorting prod_sales column
prod_sales = prod_sales.sort_values('sales', ascending=False)
```

```python
# Top 10 products by sales
prod_sales[:10]
```

Wich are the most selling products?

```python
# Grouping product name 
most_sell_product = pd.DataFrame(df.groupby('product_name').sum()['quantity'])
```

```python
# Soting most_sell_product
most_sell_product = most_sell_product.sort_values('quantity', ascending=False)
```

```python
most_sell_product[:10]
```

Which is the most preferred ship Mode?


```python
# Setting figure size
plt.figure(figsize=(10, 6))

# Plotting ship_mode
cores = ['blue', 'orange', 'green', 'red']
df['ship_mode'].value_counts().plot(kind='bar', color=cores)
plt.show()
```

Which are the most profitable category and sub-category?

```python
# Grouping category and sub category 
cat_subcat_profit = pd.DataFrame(df.groupby(['category', 'sub_category']).sum()['profit'])
```

```python
#Sorting the values
cat_subcat_profit.sort_values(['category', 'sub_category'], ascending=False)
```
